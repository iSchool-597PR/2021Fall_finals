```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.DataFrame()


list_crash_years = ['crash_09-10.csv', 'crash_10-11.csv', 'crash_11-12.csv', 'crash_12-13.csv', 'crash_13-14.csv',
                   'crash_14-15.csv', 'crash_15-16.csv', 'crash_16-17.csv', 'crash_17-18.csv', 'crash_18-19.csv']
i=1
year = 2009
dataframes_list = []
for file in list_crash_years:
    file1 = pd.read_csv(file)
    file1 = file1.drop([0,1,2,3,59,58,57])
    file1 = file1.rename(columns = {"Table 106": "States", "Unnamed: 1": "Fatalities in %s" %(year), 
                            "Unnamed: 2": "Fatalities in %s" %(year+1),
                        "Unnamed: 3": "% Change"})
    
    dataframes_list.append(file1)
    year+=1
states= dataframes_list[1]["States"].tolist()
states
dataframes_list[0].iloc[0]
```

```python
new_dict = {new_list: [] for new_list in states}
for i in range(10):
    l=0
    for item in states:
        new_dict[item].append(dataframes_list[i].iloc[l][2])
        l+=1
        
        
m=new_dict["Alabama"]
year = 2010
updated_year_dict = {}
if year<=2019:
    for values in new_dict["Alabama"]:
        updated_year_dict[year] = values
        year+=1
```
